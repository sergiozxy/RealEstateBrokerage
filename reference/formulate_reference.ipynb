```python
import pandas as pd

# Read the file
with open('sort_values.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Initialize lists to store data
bibitems = []
references = []
current_bibitem = None
current_reference = []

# Process each line
for line in lines:
    if line.startswith("\\bibitem"):
        if current_bibitem is not None:
            bibitems.append(current_bibitem)
            references.append(' '.join(current_reference))
        current_bibitem = line.strip()
        current_reference = []
    else:
        current_reference.append(line.strip())

# Add the last entry if exists
if current_bibitem is not None:
    bibitems.append(current_bibitem)
    references.append(' '.join(current_reference))

# Create a DataFrame
df = pd.DataFrame({'Bibitem': bibitems, 'Reference': references})
```

```python
df.sort_values(by='Bibitem', key=lambda x: x.str.lower(), inplace=True)
df.reset_index(drop = True, inplace = True)
```

```python
with open('sorted_values.txt', 'w', encoding='utf-8') as file:
    for index, row in df.iterrows():
        file.write(f"{row['Bibitem']}\n{row['Reference']}\n\n")
print("Sorted references have been saved to sorted_values.txt")
```
