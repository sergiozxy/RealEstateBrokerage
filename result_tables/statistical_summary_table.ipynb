# The idea

The key idea is to first convert the original name to the correct name and then store the correct name with the results, and then combine the two statistical summary to a single unit

generate another table with the name and label correspondence table

```python
data = pd.read_stata('../ttest_with_result_mean_std.dta')
data.columns = data.iloc[0]
# Drop the first row
data = data.drop(0).reset_index(drop=True)
```

```python
for i in range(1, 40):
    data.loc[i, 'SD'] = data.loc[i, 'Mean']
    data.loc[i, 'Mean'] = data.loc[i, 'Variable']
    data.loc[i, 'Variable'] = data.loc[i, 'by lj']
    data.loc[i, 'by lj'] = 0

    data.loc[40 + i, 'SD'] = data.loc[40 + i, 'Mean']
    data.loc[40 + i, 'Mean'] = data.loc[40 + i, 'Variable']
    data.loc[40 + i, 'Variable'] = data.loc[40 + i, 'by lj']
    data.loc[40 + i, 'by lj'] = 1
```

now we merge the table with the given label

```python
df = pd.read_excel('../variable__label_correspondence.xlsx')
```

```python
variable_map = {
    'income': 'income',
    'lead times': 'lead_times',
    'price conc~n': 'price_concession',
    'density': 'density',
    'broker 410': 'broker_410',
    'watching p~e': 'watching_people',
    'end price': 'end_price',
    'non online~t': 'non_online_effect',
    'watched ti~s': 'watched_times',
    'nego times': 'nego_times',
    'nego period': 'nego_period',
    'jiadian': 'jiadian',
    'kind': 'kind',
    'hotel': 'hotel',
    'shop mall': 'shop_mall',
    'museum': 'museum',
    'old': 'old',
    'ktv': 'ktv',
    'mid': 'mid',
    'prim': 'prim',
    'west food': 'west_food',
    'super': 'super',
    'sub': 'sub',
    'park': 'park',
    'area': 'area',
    'bedroom': 'bedroom',
    'toilet': 'toilet',
    'house age': 'house_age',
    'floor level': 'floor_level',
    'green ratio': 'green_ratio',
    'total buil~g': 'total_building',
    'total floo~r': 'total_floor_number',
    'living room': 'living_room',
    'elevator r~o': 'elevator_ratio',
    'kitchen': 'kitchen',
    'floor ratio': 'floor_ratio',
    'total resi~t': 'total_resident',
    'pm25': 'pm25',
    'pop': 'pop',
    'light': 'light'
}

# Replace the variable names in the 'Variable' column based on the mapping
for old_name, new_name in variable_map.items():
    data['Variable'].replace(old_name, new_name, inplace=True)
```

```python
# now we need to add some panel lines to make it compatible with the paper
lines = latex_table.split('\n')
new_line1 = '\\multicolumn{6}{l}{\\textbf{Panel 1: }Transaction property} \\\\'
new_line2 = '\\multicolumn{6}{l}{\\textbf{Panel 2: }brokerage property} \\\\'
new_line3 = '\\multicolumn{6}{l}{\\textbf{Panel 3: }hedonic information property} \\\\'
new_line4 = '\\multicolumn{6}{l}{\\textbf{Panel 4: }house property} \\\\'
new_line5 = '\\multicolumn{6}{l}{\\textbf{Panel 5: }regional property} \\\\'
# Find the index of the line containing "price_concession"
first_time = True
for i, line in enumerate(lines):
    if '\midrule' in line and first_time == True:
        # Insert the new line after the found line
        lines.insert(i + 1, new_line1)
        first_time = False
    if 'price\_concession' in line:
        # Insert the new line after the found line
        lines.insert(i + 1, new_line2)
    if 'nego\_period' in line:
        # Insert the new line after the found line
        lines.insert(i + 1, new_line3)
    if 'park' in line:
        # Insert the new line after the found line
        lines.insert(i + 1, new_line4)
    if 'floor\_ratio' in line:
        # Insert the new line after the found line
        lines.insert(i + 1, new_line5)
# Join the lines back into a single string
modified_latex_table = '\n'.join(lines)

# print(modified_latex_table)
```
